# M4 - Query RAG com Ollama

Este notebook demonstra o fluxo de retrieval + geracao. A chamada real ao Ollama so deve ser executada quando `http://localhost:11434` estiver acessivel.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'src').exists() and (PROJECT_ROOT.parent / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import requests

from src.config import settings
from src.rag import RAGService
```

```python
RUN_OLLAMA_CALL = False

try:
    response = requests.get(f"{settings.default_ollama_base_url}/api/tags", timeout=5)
    response.raise_for_status()
    ollama_available = True
    print("Ollama acessivel.")
except Exception as exc:
    ollama_available = False
    print("Ollama indisponivel neste ambiente:", exc)
    print("Defina RUN_OLLAMA_CALL = True e execute novamente quando o endpoint estiver acessivel.")
```

```python
if RUN_OLLAMA_CALL:
    rag = RAGService()
    result = rag.answer("Qual e a vigencia da convencao coletiva?")
    print(result["answer"])
    print("\nFontes:")
    for source in result["sources"]:
        print(source)
else:
    print("Execucao do RAG real desabilitada por padrao neste notebook.")
```
